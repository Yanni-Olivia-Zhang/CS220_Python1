```python
import csv

# Modified from https://automatetheboringstuff.com/chapter14/
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data
```

```python
csv_rows = process_csv("vaccinations.csv")
csv_header = csv_rows[0]
csv_data = csv_rows[1:]
```

```python
def cell(row_idx, col_name):
    col_idx = csv_header.index(col_name)
    val = csv_data[row_idx][col_idx]
    if val == "":
        return None
    return val
```

```python
#q1
dates_list = []
for i in range(len(csv_data)):
    dates_list.append(cell(i,"date"))
list(set(dates_list))
```

```python
#q2
countries_name_list = []
for i in range(len(csv_data)):
    countries_name_list.append(cell(i,"country"))
sorted(list(set(countries_name_list)))
```

```python
#q3
populations = {} #key: country name, value: popluation

for row_idx in range(len(csv_data)):
    country = cell(row_idx, "country")
    population = cell(row_idx, "population")
    if country not in populations:
        populations[country] = None # Initial place holder for country, we will replace this using next condition
    if population != None: # TODO: check that the population data is not missing
        populations[country] = int(population) # TODO: add the key-value pair to the dict
        
populations
```

```python
#q4
def daily_vaccination_date(date_wanted):
    countries_list = []
    for row_idx in range(len(csv_data)):
        date = cell(row_idx, "date")
        if date == date_wanted and cell(row_idx, "daily_vaccinations") != None:
            countries_list.append(cell(row_idx, "country"))
    return len(list(set(countries_list)))

daily_vaccination_date("10/12/2021")
```

```python
#q5
def daily_vaccinations_on(date):
    '''return a dictionary mapping each country to the number of daily vaccinations on the given date; if data is missing, the value for that country is None'''
    daily_vax_dict = {}
    for i in range(len(csv_data)):
        country_i = cell(i,"country")
        date_i = cell(i,"date")
        daily_vax_i = cell(i,"daily_vaccinations")
        if date_i == date: 
            if daily_vax_i != None:
                daily_vax_dict[country_i] = int(daily_vax_i)
            else:
                daily_vax_dict[country_i] = None# TODO: check if date_i is the correct date 
            # TODO: map the country to the number of daily vaccinations on that day
    return daily_vax_dict
daily_vaccinations_on("10/13/2021")
```

```python
#q6
def vax_across_week():
    vax_dix = {}
    for i in range (len(csv_data)):
        country = cell(i,"country")
        daily_vax = cell(i,"daily_vaccinations")
        if country not in vax_dix:
            vax_dix[country] = None
        if daily_vax != None and daily_vax != "" and vax_dix[country] == None:
            vax_dix[country] = int(daily_vax)
        elif daily_vax != None:
            vax_dix[country] += int(daily_vax)
    return vax_dix
vax_across_week()
```

```python
#q7
def find_most_vax_country():
    most_vax_num = None
    most_vax_country = None
    vax_week = vax_across_week()
    for country in vax_week:
        if vax_week[country] != None:
            if most_vax_num == None and most_vax_country == None:
                most_vax_num = vax_week[country]
                most_vax_country = country
            elif vax_week[country] > most_vax_num:
                most_vax_num = vax_week[country]
                most_vax_country = country
    return most_vax_country
find_most_vax_country()
```

```python
import datetime

def get_number_of_days(start_date, end_date):
    """Gets the number of days between the start_date (in 'mm/dd/yyyy' format) and end_date (in 'mm/dd/yyyy' format)"""
    day1 = datetime.datetime.strptime(start_date, '%m/%d/%Y') # The second argument is a format string to tell the function how to process the date string
    day2 = datetime.datetime.strptime(end_date, '%m/%d/%Y')
    delta = day2 - day1
    return delta.days
```

```python
#q10
def fully_vaccinated_by(given_date):
    '''return a dictionary mapping each country to the most recent number of people 
    fully vaccinated by the given date; if no data is available, the value is None.'''
    fully_vax_dict = {}
    for i in range(len(csv_data)):
        country = cell(i,"country")
        date = cell(i,"date")
        fully_vax = cell(i,"people_fully_vaccinated")
        if country not in fully_vax_dict:
            fully_vax_dict[country] = None
        if fully_vax == "" or fully_vax == None:
            continue
        if get_number_of_days(date, given_date) >= 0:
            fully_vax_dict[country] = int(fully_vax)
    return fully_vax_dict
fully_vaccinated_by("10/16/2021")
```

```python
#q12
vaccination_stats = {}
dates = list(set(dates_list))
countries = sorted(list(set(countries_name_list)))

def most_recent_total(given_date):
    most_recent_total = {}
    for i in range(len(csv_data)):
        country = cell(i,"country")
        date = cell(i,"date")
        total_vax = cell(i,"total_vaccinations")
        if country not in most_recent_total:
            most_recent_total[country] = None
        if total_vax == "" or total_vax == None:
            continue
        if get_number_of_days(date, given_date) >= 0:
            most_recent_total[country] = int(total_vax)
    return most_recent_total

def most_recent_people(given_date):
    most_recent_people = {}
    for i in range(len(csv_data)):
        country = cell(i,"country")
        date = cell(i,"date")
        people_vax = cell(i,"people_vaccinated")
        if country not in most_recent_people:
            most_recent_people[country] = None
        if people_vax == "" or people_vax == None:
            continue
        if get_number_of_days(date, given_date) >= 0:
            most_recent_people[country] = int(people_vax)
    return most_recent_people

for date in dates: 
    daily_vax_dict = daily_vaccinations_on(date) # a dict mapping each country to number of daily_vaccinations on date
    total_vax_dict = most_recent_total(date) # a dict mapping each country to the most recent number of total_vaccinations on date
    people_vax_dict = most_recent_people(date) # a dict mapping each country to the most recent number of people_vaccinated on date
    fully_vax_dict = fully_vaccinated_by(date) # a dict mapping each country to most recent number of people_fully_vaccinated on date
    for country in countries:
        if country not in vaccination_stats:
            vaccination_stats[country] = {}
        vaccination_stats[country][date] = {}
        vaccination_stats[country][date]['country'] = country
        vaccination_stats[country][date]['date'] = date
        vaccination_stats[country][date]['daily_vaccinations'] = daily_vax_dict[country]
        vaccination_stats[country][date]['total_vaccinations'] = total_vax_dict[country]
        vaccination_stats[country][date]['people_vaccinated'] = people_vax_dict[country]
        vaccination_stats[country][date]['people_fully_vaccinated'] = fully_vax_dict[country]
        vaccination_stats[country][date]['population'] = populations[country]
        # TODO: fill in the rest of the dict to match the example above
vaccination_stats["United States"]["10/16/2021"]
```

```python
#q17
def ratio_partly_to_fully(given_date):
    ratio_partly_to_fully = {}
    for country in vaccination_stats:
        fully = vaccination_stats[country][given_date]["people_fully_vaccinated"]
        people = vaccination_stats[country][given_date]["people_vaccinated"]
        if fully == None or people == None:
            ratio = None
        else:
            partly = people - fully
            ratio = partly / fully
        if country not in ratio_partly_to_fully:
            ratio_partly_to_fully[country] = ratio
    return ratio_partly_to_fully
ratio_partly_to_fully("10/16/2021")
```

```python
#q18
def find_lowest_ratio(given_date):
    lowest_ratio = None
    lowest_country = None
    ratio_dic = ratio_partly_to_fully(given_date)
    for country in ratio_dic:
        ratio = ratio_dic[country]
        if ratio != None:
            if lowest_ratio == None or ratio < lowest_ratio:
                lowest_ratio = ratio
                lowest_country = country
    return lowest_country
find_lowest_ratio("10/16/2021")           
```

```python
#q19
def vax_per_capita(given_date):
    vax_per_capita = {}
    for country in vaccination_stats:
        total = vaccination_stats[country][given_date]["total_vaccinations"]
        population = vaccination_stats[country][given_date]["population"]
        if total == None or population == None:
            per_capita = None
        else:
            per_capita = total / population
        if country not in vax_per_capita:
            vax_per_capita[country] = per_capita
    return vax_per_capita

def find_greatest_capita(given_date):
    greatest_capita = None
    greatest_country = None
    capita_dic = vax_per_capita(given_date)
    for country in capita_dic:
        capita = capita_dic[country]
        if capita != None:
            if greatest_capita == None or capita > greatest_capita:
                greatest_capita = capita
                greatest_country = country
    return greatest_country
find_greatest_capita("10/14/2021")
```

```python
#q20
def percentage(given_date):
    percentage = {}
    for country in vaccination_stats:
        daily = vaccination_stats[country][given_date]["daily_vaccinations"]
        population = vaccination_stats[country][given_date]["population"]
        if daily == None or population == None:
            percent = None
        else:
            percent = (daily / population) * 100
        if country not in percentage:
            percentage[country] = percent
    return percentage

def find_greatest_percentage(given_date):
    greatest_percent = None
    greatest_country = None
    percentage_dic = percentage(given_date)
    for country in percentage_dic:
        percent = percentage_dic[country]
        if percent != None:
            if greatest_percent == None or percent > greatest_percent:
                greatest_percent = percent
                greatest_country = country
    return greatest_country
find_greatest_percentage("10/10/2021")
```
